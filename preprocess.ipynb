```python
import os 
import xarray as xr
import pandas as pd
```

## ERA5

```python
ds = xr.open_dataset('era5_2023.nc')
ds.longitude.values[ds.longitude > 180] -= 360 # rearrange the longitude coordinate fom [0, 360] to [-180, 180]
ds = ds.sortby('longitude')
```

```python
import numpy as np
ds = ds.sel(longitude = np.logical_and(ds.longitude >= 80, ds.longitude < 170))
ds = ds.sel(latitude = np.logical_and(ds.latitude >= 0, ds.latitude < 50))
```

```python
ds = ds.isel(time = (pd.to_datetime("2023-01-01") <= pd.Series(ds.time.values)) & (pd.Series(ds.time.values) < pd.to_datetime("2023-02-01")))
```

```python
for month, ds_month in ds.groupby('time.month'):
    ds_month.to_netcdf(f'Cressida/ERA5/SLP/ERA5_2023{str(month).zfill(2)}.nc') # the filename should starts with variable ra in settings.py
```
